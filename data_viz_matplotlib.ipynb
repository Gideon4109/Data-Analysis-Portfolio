```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetics of Seaborn to look clean and professional
sns.set_theme(style="whitegrid")
```

```python
# Plotting a monthly trend
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_trend = [200, 240, 310, 290, 350, 420]

plt.figure(figsize=(8, 4))
plt.plot(months, sales_trend, marker='o', color='#1d3557', linewidth=2.5)
plt.title("Company Monthly Sales Volume Trend (H1 2026)", fontsize=14, pad=15)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales Volume (Units)", fontsize=12)
plt.ylim(0, 500)
plt.tight_layout()
plt.show()
```

```python
categories = ['Electronics', 'Groceries', 'Beverages', 'Home Decor']
volumes = [15000, 22000, 18500, 9500]

plt.figure(figsize=(8, 4))
sns.barplot(x=categories, y=volumes, hue=categories, palette='Blues_d', legend=False)
plt.title("Total Units Sold by Product Category", fontsize=14, pad=15)
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.tight_layout()
plt.show()
```

```python
# Comparing marketing spend to revenue
ad_spend = [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000]
revenue = [12000, 18000, 25000, 28000, 39000, 42000, 51000, 59000]

plt.figure(figsize=(8, 5))
sns.scatterplot(x=ad_spend, y=revenue, s=100, color='#e63946', alpha=0.8)
plt.title("Correlation: Marketing Ad Spend vs. Revenue ($)", fontsize=14, pad=15)
plt.xlabel("Ad Spend ($)")
plt.ylabel("Revenue ($)")
plt.tight_layout()
plt.show()
```

```python
# Customer ages
ages = [22, 25, 25, 30, 35, 36, 40, 42, 45, 50, 52, 55, 60, 65, 23, 29, 31, 33, 41, 48, 51]

plt.figure(figsize=(8, 4))
sns.histplot(ages, bins=6, kde=True, color='#4a4e69')
plt.title("Customer Age Distribution Profile", fontsize=14, pad=15)
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Compare salary distributions across departments
salaries_it = [120, 130, 135, 140, 150, 160, 250] # 250 is an outlier
salaries_hr = [80, 85, 90, 95, 100, 110, 115]

salary_data = pd.DataFrame({
    'Salary': salaries_it + salaries_hr,
    'Department': ['IT']*len(salaries_it) + ['HR']*len(salaries_hr)
})

plt.figure(figsize=(8, 4))
sns.boxplot(x='Department', y='Salary', data=salary_data, palette='Pastel1')
plt.title("Departmental Salary Distribution & Outlier Check", fontsize=14, pad=15)
plt.xlabel("Department")
plt.ylabel("Salary (Thousands $)")
plt.tight_layout()
plt.show()
```
