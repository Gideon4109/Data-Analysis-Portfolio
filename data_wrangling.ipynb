```python
import pandas as pd

# Customer information table
customers = pd.DataFrame({
    'CustomerID': [101, 102, 103, 104],
    'Name': ['Tunde', 'Chioma', 'Abubakar', 'Funmi'],
    'City': ['Lagos', 'Enugu', 'Kano', 'Ibadan']
})

# Transactions table
transactions = pd.DataFrame({
    'TransactionID': [2001, 2002, 2003],
    'CustomerID': [101, 102, 105],  # Note: Customer 105 is not in customers table
    'Amount': [15000, 42000, 8500]
})

# Performing a LEFT JOIN to keep all customer details and match transactions
left_merged = pd.merge(customers, transactions, on='CustomerID', how='left')
print("Left Merge Result:")
print(left_merged)
```

```python
# January and February Sales
jan_sales = pd.DataFrame({'Item': ['Laptop', 'Phone'], 'Revenue': [450000, 300000]})
feb_sales = pd.DataFrame({'Item': ['Tablet', 'Phone'], 'Revenue': [120000, 280000]})

# Stack them vertically (axis=0)
total_sales = pd.concat([jan_sales, feb_sales], axis=0, ignore_index=True)
print("\nVertical Concatenation:")
print(total_sales)
```

```python
# Sample sales data
sales_data = pd.DataFrame({
    'Region': ['West', 'West', 'East', 'East', 'West', 'East'],
    'Product': ['Device A', 'Device B', 'Device A', 'Device B', 'Device A', 'Device A'],
    'Revenue': [12000, 15000, 18000, 22000, 13000, 17000]
})

# Pivot to compare total revenue per product in each region
pivot_df = sales_data.pivot_table(index='Product', columns='Region', values='Revenue', aggfunc='sum')
print("\nPivot Table Result:")
print(pivot_df)
```

```python
import numpy as np

# Sample dataset with extreme outliers
prices = pd.Series([120, 150, 110, 140, 130, 1600, 95, 135, 125, -50])

Q1 = prices.quantile(0.25)
Q3 = prices.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
clean_prices = prices[(prices >= lower_bound) & (prices <= upper_bound)]
print(f"\nOriginal Prices:\n{prices.tolist()}")
print(f"Cleaned Prices (Outliers Removed):\n{clean_prices.tolist()}")
```

```python
# Sample dataframe with type errors
raw_orders = pd.DataFrame({
    'OrderID': [1, 2, 3],
    'OrderDate': ['2026-05-10', '2026-05-11', '2026-05-12'],
    'Price': ['₦65,000', '₦1,200', '₦2,500']
})

# 1. Clean the 'Price' column (remove the currency symbol and comma, then convert to float)
raw_orders['Price_Cleaned'] = raw_orders['Price'].str.replace('₦', '').str.replace(',', '').astype(float)

# 2. Parse string dates into Datetime objects
raw_orders['OrderDate_Parsed'] = pd.to_datetime(raw_orders['OrderDate'])

print("\nData Types After Cleaning:")
print(raw_orders.dtypes)
```
